## **Importing** libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import os
```

## **Load dataset**

```python
df = pd.read_csv('/content/drive/MyDrive/student-performance-dashboard-20/data/StudentPerformanceFactors.csv')


```

## **1. Histogram: Distribution of Exam Scores**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Exam_Score'], kde=True, color='blue')
plt.title('1. Histogram: Exam Score Distribution')

plt.show()
```

# **2. KDE Plot: Probability density of Study Hours**

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x='Hours_Studied', fill=True, color='orange')
plt.title('2. KDE Plot: Study Hours Density')
plt.savefig('KDE Plot.png')
plt.show()
```

### **3. Box Plot: Parental Involvement vs. Exam Score**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=df, palette='Set2')
plt.title('3. Box Plot: Parental Involvement impact on Score')
plt.show()
```

### **4. Violin Plot: School Type & Gender vs. Exam Score**

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='School_Type', y='Exam_Score', data=df, hue='Gender', split=True)
plt.title('4. Violin Plot: Score by School Type and Gender')
plt.show()
```

### **5. Scatter Plot: Attendance vs. Exam Score**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Attendance', y='Exam_Score', data=df.sample(min(1000, len(df))), hue='Motivation_Level', alpha=0.6)
plt.title('5. Scatter Plot: Attendance vs. Exam Score')
plt.show()
```

### **6. Bar Chart: Average Score by Parental Education Level**

```python
plt.figure(figsize=(8,5))
avg_edu = df.groupby('Parental_Education_Level')['Exam_Score'].mean().reset_index()
sns.barplot(x='Exam_Score', y='Parental_Education_Level', data=avg_edu, palette='viridis')
plt.title('6. Bar Chart: Avg Score by Parental Education')
plt.show()
```

### **7. Count Plot: Count of different Motivation Levels**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Motivation_Level', data=df, palette='muted', order=['Low', 'Medium', 'High'])
plt.title('7. Count Plot: Motivation Level Counts')
plt.show()
```

### **8. Pie Chart: Gender Distribution**

```python
plt.figure(figsize=(6,6))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightgreen', 'skyblue'])
plt.title('8. Pie Chart: Gender Ratio')
plt.ylabel('')
plt.show()
```

### **9. Donut Chart: Internet Access Distribution**

```python
plt.figure(figsize=(6,6))
internet = df['Internet_Access'].value_counts()
plt.pie(internet, labels=internet.index, autopct='%1.1f%%', wedgeprops=dict(width=0.4), colors=['yellow','red'])
plt.title('9. Donut Chart: Internet Access')
plt.show()
```

### **10. Heatmap: Correlation between numerical features**

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('10. Heatmap: Feature Correlation Matrix')
plt.show()
```

### **11. Line Plot: Average Score trend over Hours Studied**

```python
plt.figure(figsize=(8,5))
line_data = df.groupby('Hours_Studied')['Exam_Score'].mean().reset_index()
sns.lineplot(x='Hours_Studied', y='Exam_Score', data=line_data, marker='o')
plt.title('11. Line Plot: Mean Score vs. Study Hours')
plt.show()
```

### **12. Area Plot: Cumulative performance trend**

```python
plt.figure(figsize=(8,5))
plt.fill_between(line_data['Hours_Studied'], line_data['Exam_Score'], color="teal", alpha=0.3)
plt.plot(line_data['Hours_Studied'], line_data['Exam_Score'], color="teal")
plt.xlabel('Hours Studied')
plt.ylabel('Average Score')
plt.title('12. Area Plot: Performance Trend')
plt.show()
```

### **13. Strip Plot: Peer Influence vs. Exam Score**

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='Peer_Influence', y='Exam_Score', data=df.sample(min(500, len(df))), jitter=True, alpha=0.5)
plt.title('13. Strip Plot: Peer Influence Impact')
plt.show()
```

### **14. Swarm Plot: Family Income vs. Exam Score**

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='Family_Income', y='Exam_Score', data=df.sample(min(200, len(df))), palette='Set1')
plt.title('14. Swarm Plot: Family Income Score Distribution')
plt.show()
```

### **15. Regression Plot: Previous Scores vs. Current Exam Score**

```python
plt.figure(figsize=(8,5))
sns.regplot(x='Previous_Scores', y='Exam_Score', data=df.sample(min(500, len(df))), line_kws={'color':'green'})
plt.title('15. Regression Plot: Relationship of Previous Scores')
plt.show()
```

### 16. Point Plot: Impact of Tutoring **Sessions**

```python
plt.figure(figsize=(8,5))
sns.pointplot(x='Tutoring_Sessions', y='Exam_Score', data=df, color='green')
plt.title('16. Point Plot: Tutoring Sessions Effectiveness')
plt.show()
```

### **17. ECDF Plot: Cumulative distribution by Resource Access**

```python
plt.figure(figsize=(8,5))
sns.ecdfplot(data=df, x="Exam_Score", hue="Access_to_Resources")
plt.title('17. ECDF Plot: Exam Score by Resource Access')
plt.show()
```

### **18. Joint Plot: Hexbin distribution of Sleep vs. Score**

```python
g = sns.jointplot(x='Sleep_Hours', y='Exam_Score', data=df.sample(min(1000, len(df))), kind='hex', color='green')
g.fig.suptitle('18. Joint Plot: Sleep Hours vs. Score', y=1.02)
plt.show()
```

### **19. Stacked Bar Chart: Motivation Level by Gender**

```python
plt.figure(figsize=(8,5))
stacked = df.groupby(['Motivation_Level', 'Gender']).size().unstack()
stacked.plot(kind='bar', stacked=True, figsize=(8,5), color=['skyblue','#66b3ff'])
plt.title('19. Stacked Bar Chart: Motivation and Gender')
plt.show()
```

### **20. Boxen Plot: Learning Disabilities impact on Score**

```python
plt.figure(figsize=(8,5))
sns.boxenplot(x='Learning_Disabilities', y='Exam_Score', data=df, palette='cool')
plt.title('20. Boxen Plot: Learning Disabilities vs. Score')
plt.show()
```
